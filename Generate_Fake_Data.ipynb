```python
import os
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread, imshow, imsave
from skimage.transform import resize
```

```python
# Set location of base directory
base_path = os.path.join('/', 'workspace', 'optimization')

# Set location of datasets
images_path = os.path.join(base_path, 'images')
```

```python
# Set network settings
n_images, n_channel, n_height, n_width = 500, 3, 256, 256
```

```python
# Generate data
images = np.random.randint(low=0, high=255, size=(n_images, n_height, n_width, n_channel))
labels = np.random.randint(low=0, high=7, size=(n_images,))

# Set type
images = images.astype(np.uint8)
labels = labels.astype(np.uint8)
print('Images Shape:', images.shape, 'Images Data Type:', images.dtype)
print('Labels Shape:', labels.shape, 'Labels Data Type:', labels.dtype)
```

```python
# Show sample image
j = 10
imshow(images[j]); plt.show()
print('Label:', labels[j])
```

```python
# Save data
for i in range(n_images):
    filename = os.path.join(images_path, 'image_{:04d}_label_{}.jpg'.format(i, str(int(labels[i]))))
    #print('Saving file:', filename)
    #print('Images Shape:', images[i].shape, 'Images Data Type:', images[i].dtype)
    imsave(filename, images[i])
```

```python
# Load test images
images_file_names = os.listdir(images_path)
images_file_paths = [os.path.join(images_path, f) for f in images_file_names]
images_file_paths = [f for f in images_file_paths if f.endswith('jpg')]
images_file_paths = sorted(images_file_paths)

# Print first 2
print('Images:', images_file_paths[:2])
```

```python
 # Get the jth image
j = 0
image_file_path = images_file_paths[j]
print('Image:', image_file_path)
```

```python
# Load image
image = imread(image_file_path)

# Parse folder, image id, and label
folder, filename = os.path.split(image_file_path)
_, image_id, _, label = filename.replace('.jpg', '').split('_')
image_id, label = int(image_id), int(label)
image_id, label
print('Image {}:'.format(image_id), image.shape, image.dtype)
print('Label:', label)
```
